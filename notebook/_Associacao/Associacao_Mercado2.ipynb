```python
import pandas as pd
 
```

```python
import os.path
def path_base(base_name):
    current_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
    print(current_dir)
    data_dir = current_dir.replace('notebook','data')
    print(data_dir)
    data_base = data_dir + '\\' + base_name
    print(data_base)
    return data_base
```

```python
dados = pd.read_csv(path_base('db_mercado2.csv'),header=None)
```

```python
dados.head(3)
```

```python
transacoes = []
for i in range(0,7501):
    transacoes.append([str(dados.values[i,j]) for j in range(0,20)])
len(transacoes)
```

```python
from apyori import apriori
```

```python
#esta base tem vendas de uma semana
#supondo que tenha 4 vendas por dia >> 4*7 (1semana)
#entao 28 / 7501 = 0.0037

regras = apriori(transacoes, 
                 min_support = 0.0037,
                 min_confidence = 0.2,
                min_lift = 3,
                min_lenght = 2)
resultados = list(regras)
len(resultados)
```

```python
resultados2 = [list(x) for x in resultados]
```
